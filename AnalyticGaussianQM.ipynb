# Analytical Solution of a Stationary Gaussian Wave Packet in QM


## Definition and Normalization

From the derivation of of the free particle wave function, we have found that we can write any one-dimensional wave function $\Psi(x, t)$ as

$\Psi(x, t) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \phi(k)\exp\left[i\left(kx - \frac{\hbar k^2}{2m}t\right)\right] dk$.

Here, $k$ is the wave vector, $x$ the position, $\hbar$ the reduced Planck's constant, $m$ the mass of the particle and $t$ the time. Here $\phi(k)$ is determined from the initial condition, $\Psi(x, 0) = \psi(x)$ as
	
$\phi(k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}\psi(x) e^{-ikx} dx$.
	
In this section, we will attempt to solve for $\Psi(x, t)$ analytically assuming a Gaussian initial condition of the form
	
$\psi(x) = A\exp\left[-\frac{1}{2}\left(\frac{x - x_0}{\sigma}\right)^2\right]$,
	
where $x_0$ is the initial position of the centre of the wave packet and $\sigma$ is a parameter for the width of the wave packet. The factor of $\frac{1}{2}$ makes the solution more convenient. The constant $A$ is determined from the normalisation condition
	
$\int_{-\infty}^{\infty}|\psi(x)|^2 dx = 1$.
	
Inserting our expression for $\psi(x)$, we get

$\int_{-\infty}^{\infty}A^2 \exp\left[-\left(\frac{x - x_0}{\sigma}\right)^2\right] dx = 1$.
	
Note that the factor $\frac{1}{2}$ has dropped out because of the absolute value squared. This is simply the solution to the Gaussian integral,
	
$A^2\sqrt{\pi}\sigma = 1$.
	
Solving for $A$ gives us

$A = \frac{1}{\pi^\frac{1}{4} \sigma^\frac{1}{2}}$.
	
This gives our full expression of $\psi(x)$ as
	
$\psi(x) = \frac{1}{\pi^\frac{1}{4} \sigma^\frac{1}{2}} \exp\left[-\left(\frac{x - x_0}{\sigma}\right)^2\right]$.
	
For the moment, we will not need the normalisation constant.

## Finding $\phi(k)$

What we want to find is $\phi(k)$, which we can now determine by plugging $\psi(x)$ into the definition of $\phi(k)$,
	
$\phi(k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}\frac{1}{\pi^\frac{1}{4} \sigma^\frac{1}{2}} e^{-\frac{1}{2}\left(\frac{x - x_0}{\sigma}\right)^2} e^{-ikx} dx$.
	
We can start by taking the normalisation constant out, giving us
	
$\phi(k) = \frac{1}{\sqrt{2\pi}} \frac{1}{\pi^\frac{1}{4} \sigma^\frac{1}{2}} \int_{-\infty}^{\infty} e^{-\frac{1}{2}\left(\frac{x - x_0}{\sigma}\right)^2} e^{-ikx} dx$.
	
The constant in front of the integral does not look very convenient, but we are more interested in what happens in the integral, so let's leave them for now. We want to solve
	
$\int_{-\infty}^{\infty} e^{-\frac{1}{2}\left(\frac{x - x_0}{\sigma}\right)^2} e^{-ikx} dx$.
	
We can combine the exponentials to give us 
	
$\int_{-\infty}^{\infty} e^{-\left[\frac{1}{2}\left(\frac{x - x_0}{\sigma}\right)^2 + ikx\right]} dx$.
	
This integral can be solved by completing the square in the exponential. We want to write the exponent as
	
$a(x - b)^2 + c = ax^2 - 2abx+ab^2 + c$.
	
By expanding the square in the exponent (excluding the minus sign, as we will need it when integrating) and grouping terms with equal powers of $x$, we get
	
$\frac{1}{2}\left(\frac{x - x_0}{\sigma}\right)^2 + ikx = \frac{x^2}{2\sigma^2} - \left(\frac{x_0}{\sigma^2} - ik\right)x + \frac{x_0^2}{2\sigma^2}$.
	
Comparing this outcome with our expression above, we get three equations in three unknowns:

$a = \frac{1}{2\sigma^2}$ <br>
$2ab = \left(\frac{x_0}{\sigma^2} - ik\right)$ <br>
$ab^2 + c = \frac{x_0^2}{2\sigma^2}$. <br>
	
From this, we already have an explicit expression for $a$. We can first solve the second equation for $b$ to get
	
$b = x_0 - i\sigma^2 k$.
	
Note that $b$ has dimensions of distance, which is expected from Equation \ref{eq:expandingsquare}. We will solve the last equation by using the second equation

$ab^2 = \frac{1}{2\sigma^2}\left(x_0 - i\sigma^2 k\right)^2 = \frac{1}{2\sigma^2}\left(x_0^2 - 2i\sigma^2x_0k - \sigma^4k^2 \right) = \left(\frac{x_0^2}{2\sigma^2} - ix_0k - \frac{\sigma^2}{2}k^2 \right)$.

We can plug this back in our third equation to solve for $c$, giving us
	
$ab^2 + c = \frac{x_0^2}{2\sigma^2} - ix_0k - \frac{\sigma^2}{2}k^2 + c = \frac{x_0^2}{2\sigma^2}$.
	
The first term on the LHS cancels with the RHS term. This leaves us with an expression for $c$
	
$c = \frac{\sigma^2k^2}{2} + ix_0k$.
	
Thus, we can write the exponent in our expression as
	
$a(x - b)^2 + c$,

where $a$, $b$ and $c$ are given by
	
$a = \frac{1}{2\sigma^2}$ <br>
$b = x_0 - i\sigma^2 k$ <br>
$c = \frac{\sigma^2k^2}{2} + ix_0k$. <br>
	
The integral then becomes
	
$\int_{-\infty}^{\infty} e^{-\frac{1}{2}\left(\frac{x - x_0}{\sigma}\right)^2 -ikx} dx = \int_{-\infty}^{\infty} e^{-a(x - b)^2 - c} dx$.

We can take the $e^c$ out of the integral to give us
	
$e^{-c} \int_{-\infty}^{\infty} e^{-a(x - b)^2} dx$.
	
The solution to this integral is simply the solution to the normal Gaussian, giving
	
$e^{-c} \sqrt{\frac{\pi}{a}}$,
	
which we can express in our original variables as
	
$\sqrt{2\pi}\sigma e^{-\left(\frac{\sigma^2k^2}{2} + ix_0k\right)}$.
	
Our expression for $\phi(k)$ then becomes
	
$\phi(k) = \frac{1}{\sqrt{2\pi}} \frac{1}{\pi^\frac{1}{4} \sigma^\frac{1}{2}}\sqrt{2\pi}\sigma e^{-\left(\frac{\sigma^2k^2}{2} + ix_0k\right)}$.
	
Combining the terms in the front gives us the final expression for $\phi(k)$
	
$\phi(k) = \frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} e^{-\left(\frac{\sigma^2k^2}{2} + ix_0k\right)}$.
	
Observe that this is again a Gaussian in $k$.

## Finding $\Psi(x, t)$

We can now return to our original expression 
	
$\Psi(x, t) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \phi(k)\exp\left[i\left(kx - \frac{\hbar k^2}{2m}t\right)\right] dk$,

and substituting $\phi(k)$
	
$\Psi(x, t) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} e^{-\left(\frac{\sigma^2k^2}{2} + ix_0k\right)} e^{i\left(kx - \frac{\hbar k^2}{2m}t\right)} dk$.
	
Note that this time, we integrate to $k$ instead of $x$. We will first take the constants out of the integral without simplifying just yet.
	
$\Psi(x, t) = \frac{1}{\sqrt{2\pi}}\frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} \int_{-\infty}^{\infty}  e^{-\left(\frac{\sigma^2k^2}{2} + ix_0k\right)} e^{i\left(kx - \frac{\hbar k^2}{2m}t\right)} dk$.
	
We can then proceed by combining the exponentials
	
$\Psi(x, t) = \frac{1}{\sqrt{2\pi}}\frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} \int_{-\infty}^{\infty}  e^{-\left(\left(\frac{\sigma^2}{2} + \frac{i\hbar t}{2m} \right)k^2 + i\left(x_0 - x\right)k\right)} dk$.
	
For now, we will just focus on the integral and return to the constants later. This integral can again be solved by completing the square in the exponent. We write
	
$a(k - b)^2 + c = ak^2 - 2abk + ab^2 + c$.
	
This suggests that

$a = \frac{\sigma^2}{2} + i\frac{\hbar t}{2m}$ <br>
$2ab = -i\left(x - x_0\right)$ <br>
$ab^2 + c = 0$. <br>
	
Starting from our explicit expression for $a$, we can solve for $b$ using the second expression
	
$b = \frac{-i\left(x - x_0\right)}{2a} = \frac{-i\left(x - x_0\right)}{\sigma^2 + i\frac{\hbar t}{m}}$.
	
We will not simplify this expression any further, because it will not be of consequence to our final integral. Solving the third equation gives us
	
$c = -ab^2 = - a \left(\frac{-i\left(x - x_0\right)}{2a}\right)^2 = -\frac{-\left(x - x_0\right)^2}{4a} = \frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 + i\frac{\hbar t}{m}}$.
	
Thus, we have expressions for $a$, $b$ and $c$
	
$a = \frac{\sigma^2}{2} + i\frac{\hbar t}{2m}$ <br>
$b = \frac{-i\left(x - x_0\right)}{\sigma^2 + i\frac{\hbar t}{m}}$ <br>
$c = \frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 + i\frac{\hbar t}{m}}$
	
The integral then becomes
	
$\Psi(x, t) = \frac{1}{\sqrt{2\pi}}\frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} \int_{-\infty}^{\infty}  e^{-a(k - b)^2 - c} dk$.
	
We can again take out the $e^{-c}$, giving us
	
$\Psi(x, t) = \frac{1}{\sqrt{2\pi}}\frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}}e^{-c} \int_{-\infty}^{\infty}  e^{-a(k - b)^2} dk$.
	
The solution to the integral is again just the solution to the gaussian integral, 
	
$\Psi(x, t) = \frac{1}{\sqrt{2\pi}}\frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}}e^{-c} \sqrt{\frac{\pi}{a}}$.
	
Substituting our original expressions for $a$ and $c$, we get
	
$\Psi(x, t) = \frac{1}{\sqrt{2\pi}}\frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}}  \sqrt{\frac{\pi}{\frac{\sigma^2}{2} + i\frac{\hbar t}{2m}}} e^{-\frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 + i\frac{\hbar t}{m}}}$.
	
We can wrap this up by cleaning up some constants, giving
	
$\Psi(x, t) = \frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} \frac{1}{\sqrt{\sigma^2 + i\frac{\hbar t}{m}}} \exp\left[-\frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 + i\frac{\hbar t}{m}}\right]$.
	
This is our final wave function as a function of both position and time. Note that it reduces to $\psi(x)$ as $t = 0$.

## Probability Distribution

It would be insightful to determine the evolution of the probability distribution, rather than the wavefunction itself. We can determine this by multiplying $\Psi(x, t)$ by its complex conjugate, $\Psi^*(x, t)$.
	
$|\Psi(x, t)|^2 = \Psi(x, t)\Psi^*(x, t) = \frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} \frac{1}{\sqrt{\sigma^2 + i\frac{\hbar t}{m}}} e^{-\frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 + i\frac{\hbar t}{m}}} \cdot \frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} \frac{1}{\sqrt{\sigma^2 - i\frac{\hbar t}{m}}} e^{-\frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 - i\frac{\hbar t}{m}}}$ <br>

<br>
        
$|\Psi(x, t)|^2 = \frac{\sigma}{\sqrt{\pi}} \frac{1}{\sqrt{\sigma^4 + \left(\frac{\hbar t}{m}\right)^2}} e^{-\frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 + i\frac{\hbar t}{m}}} \cdot  e^{-\frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 - i\frac{\hbar t}{m}}}$ <br>

<br>
        
$|\Psi(x, t)|^2 = \frac{1}{\sqrt{\pi}} \frac{1}{\sqrt{\sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2}} \exp\left[{-\frac{\left(x - x_0\right)^2}{\sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2}}\right]$ <br>

	
Note again that this expression reduces to $|\psi(x)|^2$ for $t=0$. This is just a Gaussian, but with a width that depends on time, i.e.

$\sigma^\prime(t) = \sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2$
	
As time increases, the width of the Gaussian will increase as well. The probability density is still normalized:
	
$\frac{1}{\sqrt{\pi}} \frac{1}{\sqrt{\sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2}} \int_{-\infty}^{\infty}  \exp\left[{-\frac{\left(x - x_0\right)^2}{\sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2}}\right]$
<br>
$= \frac{1}{\sqrt{\pi}} \frac{1}{\sqrt{\sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2}} \sqrt{\pi\left(\sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2\right)} = 1$ <br>

## Time Scales and Plotting

In order to plot some graphs, we want to know the time scale of this broadening. We will do this by equating

$\sigma = \frac{\hbar t}{m\sigma}$.

Solving for $t$ gives us

$t = \frac{m\sigma^2}{\hbar}$.

Assuming the particle is an electron and our initial width of the Gaussian wave packet is $10^{-9}$ m, we write

$t = \frac{9.10938188 \cdot 10^{-31} \cdot 10^{-9}}{1.05457148\cdot 10^{-34}} = 8.64 \mu s$

Thus, the broadening will happen on the order of ~10 $\mu s$. Let us now plot graphs of $|\Phi(x, t)|^2$ at various times. First import some packages and define some constants.

```python
import matplotlib.pyplot as plt     # for plotting
import math                         # for pi etc.
import numpy as np                  # for numpy

x0 = 0
sigma = 1e-9
hbar = 1.05457148e-34
mass = 9.10938188e-31
```

Now let's define an array. Say that we want to see three standard deviations of the gaussian distribution, then we want $x$ to run from $x_0 - 3\sigma$ to $x_0 + 3\sigma$. We will divide this range in 1000 steps.

```python
xmin = x0 - 3*sigma
xmax = x0 + 3*sigma
plotrange = xmax - xmin
Nsteps = 1000
dx = plotrange/Nsteps

x = [xmin + jj*dx for jj in range(0, Nsteps)]
```

Now let's plot $|\Phi(x, t)|^2$ at four times

```python
T = mass * sigma**2 / hbar

T1 = 0*T
T2 = T
T3 = 2*T
T4 = 3*T
```

Now we will calculate the outputs for the four times

```python
Psi1 = np.array([0.]*Nsteps)
Psi2 = np.array([0.]*Nsteps)
Psi3 = np.array([0.]*Nsteps)
Psi4 = np.array([0.]*Nsteps)

def Gauss(T, X):
    return math.exp(-((X - x0)**2/(sigma**2 + (hbar * T / (mass* sigma))**2)))

for ii in range(0, Nsteps):
    Psi1[ii] = 1/math.sqrt(math.pi) * 1/math.sqrt(sigma**2 + (hbar * T1 / (mass * sigma))**2) * Gauss(T1, x[ii])
    
for ii in range(0, Nsteps):
    Psi2[ii] = 1/math.sqrt(math.pi) * 1/math.sqrt(sigma**2 + (hbar * T2 / (mass * sigma))**2) * Gauss(T2, x[ii])
    
for ii in range(0, Nsteps):
    Psi3[ii] = 1/math.sqrt(math.pi) * 1/math.sqrt(sigma**2 + (hbar * T3 / (mass * sigma))**2) * Gauss(T3, x[ii])
    
for ii in range(0, Nsteps):
    Psi4[ii] = 1/math.sqrt(math.pi) * 1/math.sqrt(sigma**2 + (hbar * T4 / (mass * sigma))**2) * Gauss(T4, x[ii])
```

And plot the figures

```python
ax = plt.subplot(111)                       # for the legend
ax.plot(x, Psi1, label='T1')
ax.plot(x, Psi2, label='T2')
ax.plot(x, Psi3, label='T3')
ax.plot(x, Psi4, label='T4')

plt.title('Gaussian Wave Packet')
plt.xlabel('Position [m]')
plt.ylabel('Probability')

ax.legend()

plt.grid(True)
plt.show()                                  # display plot
```
