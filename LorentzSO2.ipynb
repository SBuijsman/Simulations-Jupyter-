# The Lorentz Force and SO(2)

## Introduction

This document is first a series of two in which the Lorentz Force equation will be solved analytically for a homogeneous magnetic field. This seems trivial, but the solution will be found via an unusual approach. In this approach, a matrix formulation of the Lorentz Force equation will be determined, after which matrix exponentiation will be employed. The result will not be too surprising, but rather illuminating.

## Formulation of the problem

We will study the Lorentz Force equation,

$\mathbf{F} = q \ \mathbf{u}\times\mathbf{B}$.

Here, $\mathbf{F}$ is the Lorents Force, $q$ is the electric charge of the particle, $\mathbf{u}$ is the velocity vector of the particle and $B$ the magnetic field vector. In this text, vectors will be expressed in bold font. We will use the letter $\mathbf{u}$ for the three-dimensional velocity vector, as we will later adopt $\mathbf{v}$ as our two-dimensional velocity vector in the $xy$-plane. It is a little confusing, but we will still denote the components of $\mathbf{u}$ with $v_x$, $v_y$ and $v_z$. I just felt it would be criminal to use $\mathbf{v}$ for both the three-dimensional and the two-dimensional velocity vector. Take the magnetic field to be in the $z$-direction,

$\mathbf{B} = \begin{bmatrix}
                0 \\
                0 \\
                B
              \end{bmatrix}$,
              
and take $u$ to be in the $xy$-plane,

$\mathbf{u} = \begin{bmatrix}
                v_x \\
                v_y \\
                0
              \end{bmatrix}$.
              

## The Usual Steps

The cross product can be determined to be

$\mathbf{u}\times\mathbf{B} = \begin{bmatrix}
                                v_y B \\
                                - v_x B \\
                                0
                              \end{bmatrix}$.
                              
We can plug this back into our Lorentz Force equation to get

$\mathbf{F} = q\begin{bmatrix}
                 v_y B \\
                 - v_x B \\
                 0
               \end{bmatrix}$.
               
Given that we are considering nonrelativistic velocities, the LHS is equal to $m\dot{\mathbf{u}}$. We will write this in vector form,

$m\begin{bmatrix}
  \dot{v}_x \\
  \dot{v}_y \\
  \dot{v}_z
 \end{bmatrix} = q\begin{bmatrix}
                    v_y B \\
                    - v_x B \\
                    0
                  \end{bmatrix}$.
 
From this, it can be seen that the motion will remain in the $xy$-plane, as $\dot{v}_z$ is zero. We will leave out the uninteresting $z$-component and write

$\begin{bmatrix}
    \dot{v}_x \\
    \dot{v}_y \\
  \end{bmatrix} = \frac{qB}{m}\begin{bmatrix}
                               v_y \\
                               -v_x
                              \end{bmatrix}$.

We will now substitute

$\omega = \frac{qB}{m}$,

which is called the cyclotron frequency in literature. The expression then becomes

$\begin{bmatrix}
    \dot{v}_x \\
    \dot{v}_y \\
  \end{bmatrix} = \omega \begin{bmatrix}
                           v_y \\
                           -v_x
                         \end{bmatrix}$.

So far, the derivation has been pretty much the same as usual. From here on, we will deviate frorm the usual derivation.

## Matrix Formulation

In the usual case, we would take the two differential equations from the above equation and solve them analytically. It is well known that we will get circular orbits. In this instance, however, we will rewrite the RHS in terms of a matrix as follows

$\frac{d}{dt} \begin{bmatrix}
    v_x \\
    v_y \\
  \end{bmatrix} = \omega\begin{bmatrix}
                               0  & 1 \\
                               -1 & 0
                             \end{bmatrix} \begin{bmatrix}
                                             v_x \\
                                             v_y
                                           \end{bmatrix}$.
                                           
It is straightforward to show that this yields the same equation as above. We can now more compactly write this equation as

$\frac{d}{dt}\mathbf{v} = \omega\mathbf{A}\mathbf{v}$.

Here, $\mathbf{v}$ is now the two dimensional velocity vector in the $xy$-plane and $\bar{\mathbf{A}}$ is the matrix on the RHS of the previous equation.

## Matrix Exponentiation

Identifying

$\omega\mathbf{A}$

as an operator, the solution of the differential equation above becomes

$\mathbf{v}(t) = \mathbf{v}_0 \exp\left(\omega t\mathbf{A}\right)$.

Here, the matrix exponential is now an operator itself, which we will be attempting to find an explicit expression for.

## Finding the Matrix Exponential

To find the matrix exponential, we will write it as a Taylor-series

$\exp\left(\omega t\mathbf{A}\right) = \sum_{n=0}^\infty \frac{\left(\omega t\right)^n}{n!}\mathbf{A}^n$.

For reasons that become apparent later on, we will split the sum into even and odd terms.

$\exp\left(\omega t\mathbf{A}\right) = \sum_{n=0}^\infty \frac{\left(\omega t\right)^{2n}}{(2n)!}\mathbf{A}^{2n} + \sum_{n=0}^\infty \frac{\left(\omega t\right)^{2n+1}}{(2n+1)!}\mathbf{A}^{2n + 1}$.

The reason why this is convenient, is because we can find recurrence relations for $\mathbf{A}$.

## Recurrence Relations

Let's first look at a recurrence relation for the matrix

$\mathbf{A}^{2n} = \left(\mathbf{A}^2\right)^n$.

The definition of $\bar{\mathbf{A}}$ has been given above. We can calculate

$\left(\mathbf{A}^2\right) = -1 \begin{bmatrix}
                                  1 & 0 \\
                                  0 & 1
                                \end{bmatrix} = -\mathbf{I}$,
                                      
where $\mathbf{I}$ is the identity matrix. We can now see that

$\left(\mathbf{A}^2\right)^n = \left(-\mathbf{I}\right)^n =  \left(-1\right)^n \mathbf{I}$.

We can arrive at a similar expression for

$\mathbf{A}^{2n+1} = \mathbf{A}\mathbf{A}^{2n} = \mathbf{A}\left(\mathbf{A}^2\right)^n$.

We have found an expression for the last term already. We arrive at

$\mathbf{A}^{2n+1} = \left(-1\right)^n \mathbf{I}\mathbf{A} = \left(-1\right)^n\mathbf{A}$.

These expressions can be used to find explicit solutions for the matrix exponential.

## Final Expression for the Matrix Exponential

We have

$\exp\left(\omega t\mathbf{A}\right) = \sum_{n=0}^\infty \frac{\left(\omega t\right)^{2n}}{(2n)!}\mathbf{A}^{2n} + \sum_{n=0}^\infty \frac{\left(\omega t\right)^{2n+1}}{(2n+1)!}\mathbf{A}^{2n + 1}$.

Using the recurrence relations in the previous section, we can write

$\exp\left(\omega t\bar{\mathbf{A}}\right) = \mathbf{I}\sum_{n=0}^\infty \left(-1\right)^n \frac{\left(\omega t\right)^{2n}}{(2n)!} + \mathbf{A}\sum_{n=0}^\infty \left(-1\right)^n \frac{\left(\omega t\right)^{2n+1}}{(2n+1)!}$.

The sum over even terms can be identified as the Taylor-series for the cosine,

$\sum_{n=0}^\infty \left(-1\right)^n\frac{\left(\omega t\right)^{2n}}{(2n)!} = \cos\left(\omega t\right)$.

The sum over odd terms can be identified as the Taylor-series of the sine,

$\sum_{n=0}^\infty \left(-1\right)^n \frac{\left(\omega t\right)^{2n+1}}{(2n+1)!} = \sin\left(\omega t\right)$.

Thus,

$\exp\left(\omega t\bar{\mathbf{A}}\right) = \cos\left(\omega t\right)\mathbf{I} + \sin\left(\omega t\right)\mathbf{A}$.

Writing out the matrix in full, we get

$\exp\left(\omega t\mathbf{A}\right) = \begin{bmatrix}
                                                    \cos\left(\omega t\right)  & \sin\left(\omega t\right) \\
                                                    -\sin\left(\omega t\right) & \cos\left(\omega t\right)
                                                  \end{bmatrix}$.

This is the familiar rotation matrix in two dimensions. We will substitute

$\mathbf{R} = \begin{bmatrix}
                 \cos\left(\omega t\right)  & \sin\left(\omega t\right) \\
                 -\sin\left(\omega t\right) & \cos\left(\omega t\right)
              \end{bmatrix}$.

## The Solution

Having found the matrix exponential, we will write the solution to the Lorentz Force equation as

$\mathbf{v}(t) = \mathbf{R}\mathbf{v}_0$,

or explicitly,

$\begin{bmatrix}
    v_x(t) \\
    v_y(t) \\
  \end{bmatrix} = \begin{bmatrix}
                    \cos\left(\omega t\right)  & \sin\left(\omega t\right) \\
                    -\sin\left(\omega t\right) & \cos\left(\omega t\right)
                  \end{bmatrix} \begin{bmatrix}
                                  v_{0x} \\
                                  v_{0y}
                                \end{bmatrix}$.
                                           
This expression clearly shows that the velocity vector rotates as time progresses. The rotation matrix is a representation of an element of the Lie-Group $SO(2)$, which may represent rotations in two dimensions. It is interesting to see this group structure emerging naturally from the definition of the Lorentz Force.

## The Position

The position can be straightforwardly found from $\mathbf{v}(t)$ by integration,

$\mathbf{x}(t) = \int\mathbf{v}(t) \ dt = \left(\int\mathbf{R} \ dt\right)\mathbf{v}_0$.

The integration is only over $\mathbf{R}$, as the initial velocity vector does not depend on time. Performing the intergation of the terms inside the matrix yields

$\mathbf{x}(t) = \frac{1}{\omega} \begin{bmatrix}
                                    \sin\left(\omega t\right)  & -\cos\left(\omega t\right) \\
                                    \cos\left(\omega t\right) & \sin\left(\omega t\right)
                                  \end{bmatrix}\mathbf{v}_0 + \mathbf{c}$.
                  
The vector $\mathbf{c}$ is a constant of integration. To clear up the meaning of this constant, without loss of generality, we write the initial velocity as

$\begin{bmatrix}
   v_{0x} \\
   v_{0y}
 \end{bmatrix} = \begin{bmatrix}
                   v_0 \cos\theta_0 \\
                   v_0 \sin\theta_0
                 \end{bmatrix}$,
                 
where $\theta_0$ is some initial angle defined in a cartesian coordinate system. We can now write $\mathbf{x}(t)$ as

$\mathbf{x}(t) = \frac{v_0}{\omega}\begin{bmatrix}
                                    \sin\left(\omega t\right)  & -\cos\left(\omega t\right) \\
                                    \cos\left(\omega t\right) & \sin\left(\omega t\right)
                                  \end{bmatrix} \begin{bmatrix}
                                                  \cos\theta_0 \\
                                                  \sin\theta_0
                                                \end{bmatrix} + \mathbf{c}$.
                                                
The constant in front, in terms of our original parameters is

$\frac{v_0}{\omega} = \frac{mv_0}{qB}$.

This parameter is known as the gyroradius, the radius of the circular orbit of the particle. We will define it as

$r = \frac{mv_0}{qB}$.

Now the constant of integration can be interpreted as as the coordinate vector to the centre of the circular orbit that the particle takes.  We will call this vector $\mathbf{x}_0$ and write

$\mathbf{c} = \mathbf{x}_0$.

Written out explicitly, we get

$\begin{bmatrix}
    x(t) \\
    y(t) \\
  \end{bmatrix} = r\begin{bmatrix}
                     \sin\left(\omega t\right)  & -\cos\left(\omega t\right) \\
                     \cos\left(\omega t\right) & \sin\left(\omega t\right)
                   \end{bmatrix} \begin{bmatrix}
                                   \cos\theta_0 \\
                                   \sin\theta_0
                                 \end{bmatrix} + \begin{bmatrix}
                                                   x_0 \\
                                                   y_0
                                                 \end{bmatrix}$.
                                                                
It can be shown that $\mathbf{x}$(t) and $\mathbf{v}$(t) are perpendicular, up to the integration constant $\mathbf{x}_0$, which is expected. In the next document, general expressions for three-dimensional motion will be derived in a similar manner.

## NB: Relativity

The relativistic expression of the Lorentz-Force without electric fields reads

$\frac{d(\gamma m \mathbf{u})}{dt} = q\mathbf{u}\times\mathbf{B}$.

Here, $\gamma$ is the relativistic gamma factor. One would think that, given that this is a function of $t$, this would complicate matters. However, since magnetic fields cannot perform work, the particle's speed (magnitude of the velocity vector) never increases. Since $\gamma$ only depends on this speed, it is constant in time. The expression can then be written as

$\frac{d\mathbf{u}}{dt} = \frac{q}{\gamma m}\mathbf{u}\times\mathbf{B}$.

All that changes to our cyclotron frequency and gyroradius is

$\omega \rightarrow \frac{\omega}{\gamma}$,

and

$r \rightarrow \gamma r$.
