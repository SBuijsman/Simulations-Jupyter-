# Polynomial Solver (Degree 4 and Less)

## Introduction
Several problems require calculating the roots of a polynomial. One example of such a problem is solving a ordinary linear homogeneous differential equation. In this document, we will develop a general program that can be used to calculate the roots of polynomials up to order 4 exactly. This program was developed during the creation of the analytic solution of a third order linear homogeneous ordinary differential equation and is meant to divert the focus from the solution of a third degree polynomial. In this way, the problem at hand can receive full attention, with only a reference to this document.

This document will provide four functions, one for each polynomial of degree 1 through 4. Since a general formula in terms of simple functions cannot exist for polynomials of degree 5 and higher, such polynomials will not be considered here. The development of the functions will increase in complexity with each degree. A polynomial of degree $n$ also has $n$ roots. These roots may in general be complex numbers. The functions will expect the coefficients of the polynomial and will return an array with the roots. The classification of whether the roots are real or complex will not be explicitly returned.

## Polynomial of Degree 1 (Linear)

A polynomial of degree 1 has the general form

$f(x) = ax + b$.

Here, $a$ and $b$ are constants that are, in general, complex numbers. It is our goal to find the roots of this polynomial, i.e.

$f(x) = 0$, <br>
$ax + b = 0$.

It is apparent that the root of this equation is

$x = -\frac{b}{a}$,

Given that $a \neq 0$. It can be seen that $x \in \mathbb{R}$ if either both $a$ and $b$ are real, or if both are purely imaginary. If either $a$ or $b$ are purely imaginary, $x$ will be purely imaginary. If either or both $a$ and $b$ are complex (i.e., have a real and an imaginary component), $x$ will be complex too. Again, we will not be too concerned with these facts, but it is good to keep track of the possible results. Below, a simple function is defined to calculate the roots of this polynomial.

```python
def RootLinear(a, b):
    
    # Checks if a is not zero. Since it will not be in most cases, this statement comes first.
    if a != 0:
        return -a/b
    
    # If a = 0, an error message will be returned.
    else:
        return "Cannot Compute, a = 0."
```

The function will first check if $a \neq 0$. If it is not (which will be most cases), it will compute the root. If $a = 0$, it will return an error message. Below are six tests of this function for various inputs of $a$ and $b$.

```python
# Tests:

#-------------------------------------------------------------------------------
# First case: Both a and b real, a nonzero.
print("Both a and b real.")
a = 5
b = 2

print(f"a = {a}.")
print(f"b = {b}.")
print(f"x = {RootLinear(a, b)}\n")

#-------------------------------------------------------------------------------
# Second case: Both a and b imaginary, a nonzero.
print("Both a and b imaginary.")
a = complex(0, 10)
b = complex(0, 3)

print(f"a = {a}.")
print(f"b = {b}.")
print(f"x = {RootLinear(a, b)}\n")

#-------------------------------------------------------------------------------
# Third case: a imaginary, b real, a nonzero.
print("a real, b imaginary")
a = complex(7, 0)
b = complex(0, 9)

print(f"a = {a}.")
print(f"b = {b}.")
print(f"x = {RootLinear(a, b)}\n")

#-------------------------------------------------------------------------------
# Fourth case: a real, b imaginary, a nonzero.
print("a imaginary, b real.")
a = complex(0, 2)
b = complex(8, 0)

print(f"a = {a}.")
print(f"b = {b}.")
print(f"x = {RootLinear(a, b)}\n")

#-------------------------------------------------------------------------------
# Fifth case: Both a and b complex, a nonzero.
print("Both a and b complex.")
a = complex(5, 2)
b = complex(8, 4)

print(f"a = {a}.")
print(f"b = {b}.")
print(f"x = {RootLinear(a, b)}\n")

#-------------------------------------------------------------------------------
# Sixth case: b complex, a zero.
print("a zero, b complex.")
a = 0
b = complex(8, 7)

print(f"a = {a}.")
print(f"b = {b}.")
print(f"x = {RootLinear(a, b)}\n")
```

We can see that the function works as desired. We will now proceed with finding roots of quadratic functions.

## Polynomial of Degree 2 (Quadratic)

A polynomial of degree 2 has the general form

$f(x) = ax^2 + bx + c$.

Here, $a$, $b$ and $c$ may again be complex constants. To find the roots of this polynomial, it is required to solve for $x$ the equation

$f(x) = 0$, <br>
$ax^2 + bx + c = 0$.

It is well known that this can be done by the quadratic formula, given by

$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$.

There are again many cases in which $x$ can be purely real/imaginary or have both a real and imaginary component. We will not investigate all cases here. Below is a definition of the function. Again, the case where $a = 0$ is special. We will not get an error message this time, however. If $a = 0$, we will just have to solve a linear equation with the two remaining coeffictions $b$ and $c$. Thus, the function will first check if $a \neq 0$, after which it will apply the quadratic formula and return an array with two elements; the two roots of the polynomial. If $a = 0$, it will call the `RootLinear` function, defined in the previous section. Its arguments will then be $b$ and $c$. The function will now display a message, clarifying that $a = 0$ and that we are actually solving a linear function. The two elements of the return array will therefore be identical.

```python
def RootQuadratic(a, b, c):
    if a != 0:
        return [(-b + (b**2 - 4*a*c)**(1/2))/(2*a), (-b - (b**2 - 4*a*c)**(1/2))/(2*a)]
    
    if a == 0:
        print("Note: Function is linear, not quadratic.")
        Root = RootLinear(b, c)
        return [Root, Root]
```

Below, several test cases are outlined.

```python
# Tests:

#-------------------------------------------------------------------------------
# First case: a, b and c real, a nonzero.
print("a, b and c real.")
a = 5
b = 2
c = 2

print(f"a = {a}.")
print(f"b = {b}.")
print(f"c = {c}.")
print(f"x = {RootQuadratic(a, b, c)}\n")

#-------------------------------------------------------------------------------
# Second case: a, b and c complex, a nonzero.
print("a, b and c complex")
a = complex(1, 5)
b = complex(2, 7)
c = complex(9, 0)

print(f"a = {a}.")
print(f"b = {b}.")
print(f"c = {c}.")
print(f"x = {RootQuadratic(a, b, c)}\n")

#-------------------------------------------------------------------------------
# Third case: a zero, b and c complex.
print("a zero, b and c complex")
a = 0
b = complex(1, 9)
c = complex(7, 3)

print(f"a = {a}.")
print(f"b = {b}.")
print(f"c = {c}.")
print(f"x = {RootQuadratic(a, b, c)}\n")

#-------------------------------------------------------------------------------
# Fourth Case: Discriminant zero
print("Discriminant zero.")
a = 1
b = 4
c = 4

print(f"a = {a}.")
print(f"b = {b}.")
print(f"c = {c}.")
print(f"x = {RootQuadratic(a, b, c)}\n")

#-------------------------------------------------------------------------------
# Fifth case: Discriminant real and negative.
print("Discriminant real and negative.")
a = 1
b = 2
c = 4

print(f"a = {a}.")
print(f"b = {b}.")
print(f"b = {c}.")
print(f"x = {RootQuadratic(a, b, c)}\n")
```

The function works as expected in the cases outlined above. Now with the relatively simple linear and quadratic cases out of the way, we will proceed with the rather involved cubic formula.

## Polynomial of Degree 3 (Cubic)

A polynomial of degree 3 has the general form

$f(x) = ax^3 + bx^2 + cx + d$.

We would like to find the roots such that

$f(x) = 0$, <br>
$ax^3 + bx^2 + cx + d = 0$.

This can be done by a less well-known procedure, which is rather involved. The steps will be outlined in this section. Most of the understanding comes from the video below.

### Depressed Cubic

The polynomial in $x$ can be simplified by noting that any cubic polynomial is antisymmetric around its point of inflection. It useful to find this point and changing variables such that the point of inflection lies at the zero point of this changed coordinate system. This will be demonstrated below.

To find the point of inflection, we will compute the second derivative of the polynomial and equate it to zero. This is

$6ax + 2b = 0$.

The point of inflection thus lies at

$x = -\frac{b}{3a}$.

We can translate $x$ to get

$x = y - \frac{b}{3a}$.

We can do this, as long as we remember that we need to substitute $y = x + \frac{b}{3a}$ back in at the end. Plugging our expression with $y$ into our polynomial, we get

$a\left(y - \frac{b}{3a}\right)^3 + b\left(y - \frac{b}{3a}\right)^2 + c\left(y - \frac{b}{3a}\right) + d = 0$.

We can expand the brackets to give

$a \left(y^3 - 3\frac{b}{3a}y^2 + 3\left(\frac{b}{3a}\right)^2 y - \left(\frac{b}{3a}\right)^3\right) + b \left(y^2 - 2\frac{b}{3a}y + \left(\frac{b}{3a}\right)^2\right) + c\left(y - \frac{b}{3a}\right) + d = 0$.

Note that the terms with $y^2$ cancel. We can group terms with equal powers in $y$ to get

$ay^3 + \left(c - \frac{b^2}{3a}\right) y + \left(\frac{2b^3}{27a^2} -\frac{bc}{3a} + d\right) = 0$

As a final step, we can divide by $a$ to free up the $y^3$ term. This gives us

$y^3 + \left(\frac{c}{a} - \frac{b^2}{3a^2}\right) y + \left(\frac{2b^3}{27a^3} -\frac{bc}{3a^2} + \frac{d}{a}\right) = 0$

Now this hardly looks simpler. However, we can make the substitutions

$p = \frac{c}{a} - \frac{b^2}{3a^2}$ <br>
$q = \frac{2b^3}{27a^3} -\frac{bc}{3a^2} + \frac{d}{a}$.

This reduces the above equation to

$y^3 + py + q = 0$.

This is the so-called depressed cubic. Any cubic equation can be reduced to a depressed cubic, so up until now there is no loss of generality. Defining values for $p$ and $q$ will be the first required element in our function for finding roots for cubic polynomials.


### Formula for Finding $y$

Having formulated the problem as a depressed cubic, we can use the general formula for finding cube roots,

$y = \sqrt[3]{-\frac{q}{2} + \sqrt{\left(\frac{q}{2}\right)^2 + \left(\frac{p}{3}\right)^3}} + \sqrt[3]{-\frac{q}{2} - \sqrt{\left(\frac{q}{2}\right)^2 + \left(\frac{p}{3}\right)^3}}$.

We will not derive this result here. The most important fact about this formula is the term under the square root. We will define it as

$D = \left(\frac{q}{2}\right)^2 + \left(\frac{p}{3}\right)^3$.

This is the discriminant for cubic equations and it defines the nature of the solutions for $y$. There are three distinct cases and one exception to these cases. Defining the discriminant will be the second required element in our function for finding roots for cubic polynomials. We will first consider three cases for $D$. In all these cases, $p$, $q$ and thus $D$ are assumed to be real. We can then make the distinction in which $D < 0$, $D = 0$ and $D > 0$. Later we will consider the case when $p$ and $q$ can be complex.


### Case 1: $D < 0$

Even though we are taking square roots of a negative number, the roots will turn out to be purely real. This surprising result will be shown here. We write down 

$y = \sqrt[3]{-\frac{q}{2} + \sqrt{D}} + \sqrt[3]{-\frac{q}{2} - \sqrt{D}}$.

This is just the general formula for finding cube roots, but with the substitution for the discriminant. If $D \in \mathbb{R}$ and $D < 0$, then $\sqrt{D}$ is purely imaginary and we can write

$\sqrt{D} = i\sqrt{|D|}$,

where $|D|$ is the absolute value of $D$. The roots of $y$ then become

$y = \sqrt[3]{-\frac{q}{2} + i\sqrt{|D|}} + \sqrt[3]{\frac{q}{2} - i\sqrt{|D|}}$.

We can now express the number under the cube root in exponential form,

$\frac{q}{2} + i\sqrt{|D|} = \rho e^{i\phi}$, <br>
$\frac{q}{2} - i\sqrt{|D|} = \rho e^{-i\phi}$,

where

$\rho = \sqrt{\left(\frac{q}{2}\right)^2 + |D|}$, <br>
$\phi = \arccos\left(\frac{q/2}{\rho}\right)$.

This is a rather unconventional to define $\phi$, but it allows us to find a purely real expression for $y$ in the end. We can simplify $\rho$ a bit further by noting that

$|D| = -D = -\left(\left(\frac{q}{2}\right)^2 + \left(\frac{p}{3}\right)^3\right)$.

Plugging this back into our expression for $\rho$, we get

$\rho = \sqrt{\left(\frac{q}{2}\right)^2 - \left(\left(\frac{q}{2}\right)^2 + \left(\frac{p}{3}\right)^3\right)} = \left(-\frac{p}{3}\right)^{\frac{3}{2}}$.

This in turn allows us to write

$\phi = \arccos\left(\frac{3q}{2p} \sqrt{-\frac{3}{p}}\right)$.

Plugging in the exponential form into the expression for our roots, we get

$y = \sqrt[3]{\rho e^{-i\phi}} + \sqrt[3]{\rho e^{i\phi}}$.

Taking the cube root of this expression is very simple. We get three distinct roots, given by

$y_k = \rho^{1/3}\left(\exp\left[i\left(\frac{\phi}{3} + \frac{2\pi k}{3}\right)\right] + \exp\left[-i\left(\frac{\phi}{3} + \frac{2\pi k}{3}\right)\right]\right)$.

Here, we denoted $y$ with a subscript $k$ for the three distinct values it can take, where $k \in \{0, 1, 2\}$. Roots with different values for $k$ are not distinct from these three cases. The term between brackets is just the cosine of the argument, giving

$y_k = 2\rho^{1/3}\cos\left(\frac{\phi}{3} + \frac{2\pi k}{3}\right)$.

We can now plug our expressions for $\rho$ and $\phi$ back in to get

$y_k = 2\sqrt{-\frac{p}{3}}\cos\left(\frac{1}{3}\arccos\left(\frac{3q}{2p} \sqrt{-\frac{3}{p}}\right) + \frac{2\pi k}{3}\right)$.

The square root term in front of the cosine might be a bit confusing, but note that $D < 0$. The definition of $D$ is

$D = \left(\frac{q}{2}\right)^2 + \left(\frac{p}{3}\right)^3$.

Since the squared term cannot be negative, $p$ has to be negative in order to make $D < 0$. This means that the $\sqrt{-p/3}$ is actually the square root of a positive number and is therefore real. Since the argument of the cosine and arccosine are all real, the three roots are also real.

To conclude, our function needs to return an array with elements $y_k$ for $k \in \{0, 1, 2\}$ according to the expression above, in the case when $D < 0$.

### Case 2: $D > 0$

In the case where $D > 0$, the equation has one real solution and thus two complex ones. We can again write 

$y_1 = \sqrt[3]{-\frac{q}{2} + \sqrt{D}} + \sqrt[3]{-\frac{q}{2} - \sqrt{D}}$.

Since $D > 0$, the square roots are real and we are taking the cube root of a real number. We call it $y_1$, because it will be our first root. It turns out that the two complex roots are complex conjugates of each other and can be obtained by multiplying the real root by a cube root of one, i.e.

$\sqrt[3]{1} = e^\frac{2i\pi k}{3}$,

where $k \in \{0, 1, 2\}$.

So we can write the other two roots as

$y_2 = y_1e^\frac{2i\pi}{3}$, <br>
$y_3 = y_1e^\frac{4i\pi}{3}$.

Or, in general

$y_k = e^\frac{2i\pi k}{3}\left(\sqrt[3]{-\frac{q}{2} + \sqrt{D}} + \sqrt[3]{-\frac{q}{2} - \sqrt{D}}\right)$.


### Case 3: $D = 0$

When $D = 0$, the polynomial has only two distinct roots. They are

$y_1 = -2\left(\frac{q}{2}\right)^\frac{1}{3}$, <br>
$y_2 = y_3 = \left(\frac{q}{2}\right)^\frac{1}{3}$.

This can be directly shown by writing

$t^3 + pt + q = \left(t + 2\left(\frac{q}{2}\right)^\frac{1}{3}\right)\left(t - \left(\frac{q}{2}\right)^\frac{1}{3}\right)^2$.

Expanding the brackets yields

$t^3 - 3\left(\frac{q}{2}\right)^\frac{2}{3}t + q$.

Now, since $D = 0$, 

$\left(\frac{q}{2}\right)^2 + \left(\frac{p}{3}\right)^3 = 0$,

and

$-3\left(\frac{q}{2}\right)^\frac{2}{3} = p$.

This brings the expanded expression back to our original depressed cubic. Note that these roots work equally well when $p = q = 0$, so that case does not require additional attention.


### Case 4: $q, D \in \mathbb{C}$

In the case when either or both $q$ and $D$ are complex (i.e. have a real and imaginary part), we will apply the same formula as above,

$y = \sqrt[3]{-\frac{q}{2} + \sqrt{D}} + \sqrt[3]{-\frac{q}{2} - \sqrt{D}}$.

We would like to find a simple and unambiguous expression for the roots. In order to make this calculation manageable, we will make quite a few substitutions from now on. We first write both $q$ and $D$ as complex numbers, i.e.

$q = a + bi$, <br>
$D = c + di$,

where $a, b, c, d \in \mathbb{R}$. We can then first express $D$ in exponential form as

$D = \rho e^{i\phi}$,

where

$\rho = \sqrt{c^2 + d^2}$, <br>
$\phi = \arctan\left(\frac{d}{c}\right)$.

The square root of $D$ is then given by

$\sqrt{D} = \rho^\frac{1}{2}e^{i\frac{\phi}{2}}$.

We can rewrite this expression in terms of the original parameters $c$ and $d$ as

$\sqrt{D} = \left(c^2 + d^2\right)^\frac{1}{4} \left(\cos\left(\frac{1}{2}\arctan\left(\frac{d}{c}\right)\right) + i\sin\left(\frac{1}{2}\arctan\left(\frac{d}{c}\right)\right)\right)$.

This expression is rather unwieldy. We can simplify it by defining some new constants,

$c^\prime = \left(c^2 + d^2\right)^\frac{1}{4} \cos\left(\frac{1}{2}\arctan\left(\frac{d}{c}\right)\right)$ <br>
$d^\prime = \left(c^2 + d^2\right)^\frac{1}{4} \sin\left(\frac{1}{2}\arctan\left(\frac{d}{c}\right)\right)$,

such that $\sqrt{D}$ becomes

$\sqrt{D} = c^\prime + d^\prime i$.

We now want to arrive at an exponential expression for the term under the cube roots. We will denote this by

$u_\pm = -\frac{q}{2} \pm \sqrt{D}$.

The root then becomes

$y = \sqrt[3]{u_+} + \sqrt[3]{u_-}$.

We can write

$u_\pm = -\frac{a}{2} - \frac{b}{2}i \pm \left(c^\prime + d^\prime i\right)$.

Grouping terms with and without $i$ gives us

$u_\pm = \left(-\frac{a}{2} \pm c^\prime\right) + \left(-\frac{b}{2} \pm d^\prime\right)i$.

We can again redefine some constants as

$A_\pm = -\frac{a}{2} \pm c^\prime$, <br>
$B_\pm = -\frac{b}{2} \pm d^\prime$.

Then

$u_\pm = A_\pm + B_\pm i$.

We can rewrite this in exponential form as

$u_\pm = \rho_\pm e^{i\phi_\pm}$,

where

$\rho_\pm = \sqrt{A_\pm^2 + B_\pm^2}$, <br>
$\phi_\pm = \arctan\left(\frac{B_\pm}{A_\pm}\right)$.

In terms of our original parameters, these are

$\rho_\pm = \sqrt{\left(-\frac{a}{2} \pm \left(c^2 + d^2\right)^\frac{1}{4} \cos\left(\frac{1}{2}\arctan\left(\frac{d}{c}\right)\right)\right)^2 + \left(-\frac{b}{2} \pm \left(c^2 + d^2\right)^\frac{1}{4} \sin\left(\frac{1}{2}\arctan\left(\frac{d}{c}\right)\right)\right)^2}$, <br>
$\phi_\pm = \arctan\left(\frac{-\frac{b}{2} \pm \left(c^2 + d^2\right)^\frac{1}{4} \sin\left(\frac{1}{2}\arctan\left(\frac{d}{c}\right)\right)}{-\frac{a}{2} \pm \left(c^2 + d^2\right)^\frac{1}{4} \cos\left(\frac{1}{2}\arctan\left(\frac{d}{c}\right)\right)}\right)$.

Of course, these expressions are absolutely opaque. It will serve our purpose, however. We can let the function first calculate $\rho_\pm$ and $\phi_\pm$. The root is then easily calculated as

$y = \sqrt[3]{u_+} + \sqrt[3]{u_-} = \rho_+^\frac{1}{3}e^{i\frac{\phi_+}{3}} + \rho_-^\frac{1}{3}e^{i\frac{\phi_-}{3}}$.

Again, this root is one of three. The other roots are given by multiplying this one by a factor of $e^{i\frac{2\pi}{3}}$ and $e^{i\frac{4\pi}{3}}$. Thus,

$y_k = \left(\rho_+^\frac{1}{3}e^{i\frac{\phi_+}{3}} + \rho_-^\frac{1}{3}e^{i\frac{\phi_-}{3}}\right)e^{i\frac{2k\pi}{3}}$,

where $k \in \{0, 1, 2\}$.

Our function will classify whether $q$ and/or $D$ are complex by comparing its imaginary part to zero. If both imaginary components are zero, the above formula will be applied. At the stages where the arctangents needs to be calculated, it will be checked whether the denominator is zero. In this case, it will yield $\pm\frac{\pi}{2}$. Since in most cases to which this function will be applied, the parameters will not be complex, this feature is placed at the end.

```python
# math package is required for trig functions.
import math

# Since Python sometimes has issures calculating cube roots, we will define a function for that.
def get_cuberoot(x):
    if x < 0:
        x = abs(x)
        cube_root = x**(1/3)*(-1)
    else:
        cube_root = x**(1/3)
    return cube_root

# The actual function.
# Expects: For real or complex inputs.
# Returns: An array of three real or complex values.
def RootCubic(a, b, c, d):
    
    # If a = 0, we're actually solving a quadratic. If not, the body of the function is initiated.
    if a != 0:
        # Define p, q and D.
        p = c/a - 3*b**2/a**2
        q = 2*b**3/(27*a**3) - b*c/(3*a**2) + d/a
        
        D = (q/2)**2 + (p/3)**3
        
        # Print them, as an extra check.
        print(f"D = {D}")
        print(f"p = {p}")
        print(f"q = {q}\n")
        
        # If p, q and D are all real, we can proceed with comparing the discriminant with zero and act accordingly.
        if D.imag == 0 and q.imag == 0 and p.imag == 0:
        
            # See section "Case 1: D < 0" for further explanation.
            # Note that we add b/3a to correct for our conversion from x to y.
            if D < 0:
                root1 = 2*(-p/3)**(1/2) * math.cos(1/3 * math.acos(3*q/(2*p) * (-p/3)**(1/2))) + b/(3*a)
                root2 = 2*(-p/3)**(1/2) * math.cos(1/3 * math.acos(3*q/(2*p) * (-p/3)**(1/2)) + 2*math.pi/3) + b/(3*a)
                root3 = 2*(-p/3)**(1/2) * math.cos(1/3 * math.acos(3*q/(2*p) * (-p/3)**(1/2)) + 4*math.pi/3) + b/(3*a)
            
                return [root1, root2, root3]

            # See section "Case 2: D > 0" for further explanation.
            # Note that we add b/3a to correct for our conversion from x to y.
            if D > 0:
                root1 = get_cuberoot(-q/2 + D**(1/2)) + get_cuberoot(-q/2 - D**(1/2))
                root2 = 1/2 * complex(-root1, 3**(1/2) * root1) + b/(3*a)
                root3 = 1/2 * complex(-root1, -3**(1/2) * root1) + b/(3*a)
            
                return [root1 + b/(3*a), root2, root3]

            # See section "Case 3: D = 0" for further explanation.
            # Note that we add b/3a to correct for our conversion from x to y.
            if D == 0:
                root1 = -2*(q/2)**(1/3) + b/(3*a)
                root2 = root3 = (q/2)**(1/3) + b/(3*a)
            
                return [root1, root2, root3]
        
        # See section "Case 4: D q, D in C" for further explanation.
        # Note that we add b/3a to correct for our conversion from x to y.
        else:
            a = q.real
            b = q.imag
            c = D.real
            d = D.imag
            
            rhop = ((-a/2 + (c**2 + d**2)**(1/4) * math.cos(1/2 * math.atan(d/c)))**2 + (-b/2 + (c**2 + d**2)**(1/4) * math.sin(1/2 * math.atan(d/c)))**2)**(1/2)
            rhom = ((-a/2 - (c**2 + d**2)**(1/4) * math.cos(1/2 * math.atan(d/c)))**2 + (-b/2 - (c**2 + d**2)**(1/4) * math.sin(1/2 * math.atan(d/c)))**2)**(1/2)
            phip = math.atan((-b/2 + (c**2 + d**2)**(1/4) * math.sin(1/2 * math.atan(d/c))) / (-a/2 + (c**2 + d**2)**(1/4) * math.cos(1/2 * math.atan(d/c))))
            phim = math.atan((-b/2 - (c**2 + d**2)**(1/4) * math.sin(1/2 * math.atan(d/c))) / (-a/2 - (c**2 + d**2)**(1/4) * math.cos(1/2 * math.atan(d/c))))
                             
            root1 = get_cuberoot(rhop) * complex(math.cos(phip/3), math.sin(phip/3)) + get_cuberoot(rhom) * complex(math.cos(phim/3), math.sin(phim/3))
            root2 = 1/2 * complex(-root1, 3**(1/2) * root1) + b/(3*a)
            root3 = 1/2 * complex(-root1, -3**(1/2) * root1) + b/(3*a)
                             
            return [root1 + b/(3*a), root2, root3]
        
    else:
        print("Note: Function is quadratic, not cubic.")
        return RootQuadratic(b, c, d)
```
