# Gallilean Covariance of Quantum Mechanical Wave Functions

## Introduction

In a previous document, the time evolution of a Gaussian quantum mechanical wave packet was studied. We only considered a stationary wave, representing a particle that is at rest. The Gaussian wave function was found to be 

$\Psi(x, t) = \frac{\sigma^\frac{1}{2}}{\pi^\frac{1}{4}} \frac{1}{\sqrt{\sigma^2 + i\frac{\hbar t}{m}}} \exp\left[-\frac{1}{2}\frac{\left(x - x_0\right)^2}{\sigma^2 + i\frac{\hbar t}{m}}\right]$.

The corresponding probability density is

$|\Psi(x, t)|^2 = \frac{1}{\sqrt{\pi}} \frac{1}{\sqrt{\sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2}} \exp\left[{-\frac{\left(x - x_0\right)^2}{\sigma^2 + \left(\frac{\hbar t}{m\sigma}\right)^2}}\right]$.

It was noted that this is a Gaussian distribution, with a standard deviation depending on time. This illuminates how the state spreads out as time progresses. In this document we will generalise the free Gaussian wave packet by studing the covariance of quantum mechanical states under Gallilean boosts.

## Transformed Coordinates

Suppose that we are in a stationary one-dimensional frame of reference parametrised by the coordinate $x$ and the time $t$. Now suppose that an observer passes us at constant velocity $v$. Their coordinates in our frame of reference may be expressed as

$X = x - vt$, <br>
$T = t$.

We consider $v$ to be small compared to the speed of light such that we may neglect relativistic effects (in which case we could no longer study this problem only in terms of classical quantum mechanics, but would require a quantum field theoretical framework).

## Moving State

Now suppose that the observer moving relative to us has prepared a particle in a Gaussian state as given above. This state is stationary with respect to their own frame of reference, meaning that the particle is at rest in their frame. In our frame, the particle is moving. We call the state in their frame $\Psi^\prime(X, T)$. We would like to investigate how the moving state corresponds to the stationary state in the other frame, i.e.:

$\Psi(x, t) \Longleftrightarrow \Psi^\prime(X, T)$.

In order to do this, we start by observing that no matter the reference frame, both states must remain normalised,

$\int_{-\infty}^\infty |\Psi(x, t)|^2 dx = \int_{-\infty}^\infty |\Psi^\prime(X, T)|^2 dx^\prime = 1$

This implies that 

$|\Psi(x, t)|^2 = |\Psi^\prime(X, T)|^2$.

This does not imply that the states must be equal, but it narrows down our search a lot. Now we know that the states must be equal up to a complex phase

$\Psi(x, t) = \Psi^\prime(X, T) e^{if(X, T)}$.

In this equation, $f(X, T)$ is some function of both $X$ and $T$. Its value must be real in order for the exponent to remain completely imaginary. If this were not the case, the state could not remain normalised, as there would be a real exponent causing exponential growth/decline.

## Finding $f(X, T)$
### Transforming the Schrödinger Equation

Our current goal is to find this function $f(X, T)$. To this end we will employ the Schrödinger Equation. We will start by writing down the Schrödinger Equation in our own frame of reference, 

$i\hbar \frac{\partial \Psi}{\partial t} = - \frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + V(x)\Psi$.

Note that we have written down the potential $V(x)$, even though we are considering a free particle. The potential is included in order to show that the transformation we will find also applies to non-free particles. We will start immediately by simplifying the LHS by dividing through $i\hbar$,

$\frac{\partial \Psi}{\partial t} = i\frac{\hbar}{2m} \frac{\partial^2 \Psi}{\partial x^2} + \frac{V(x)}{i\hbar}\Psi$.

We will rename our potential, 

$\frac{V(x)}{i\hbar} = W(x)$,

reducing the Schrödinger Equation to

$\frac{\partial \Psi}{\partial t} = i\frac{\hbar}{2m} \frac{\partial^2 \Psi}{\partial x^2} + W(x)\Psi$.

We will now boost this equation to the moving frame. To this end, we must transform the derivatives. This is done as follows,

$\frac{\partial}{\partial x} = \frac{\partial}{\partial X} \frac{\partial X}{\partial x} + \frac{\partial}{\partial T} \frac{\partial T}{\partial x} = \frac{\partial}{\partial X}$ <br>

$\frac{\partial}{\partial t} = \frac{\partial}{\partial X} \frac{\partial X}{\partial t} + \frac{\partial}{\partial T} \frac{\partial T}{\partial t} = -v \frac{\partial}{\partial X} + \frac{\partial}{\partial T}$ <br>

We can replace the derivatives in our original Schrödinger equation, as well as the state by the transformed one. This is

$\left(-v \frac{\partial}{\partial X} + \frac{\partial}{\partial T}\right) \left(\Psi^\prime (X, T) e^{if(X, T)}\right) = i\frac{\hbar}{2m} \frac{\partial^2}{\partial X^2}\left(\Psi^\prime (X, T) e^{if(X, T)}\right) + W^\prime(X, T)\left(\Psi^\prime (X, T) e^{if(X, T)}\right)$.

Here, $W^\prime(X, T)$ is the potential, transformed to the new coordinates. For compactness, we will omit the function arguments. We just need to remember that $\Psi^\prime = \Psi^\prime(X, T)$, $W^\prime = W^\prime(X, T)$ and $f = f(X, T)$. This reduces the above equation to

$\left(-v \frac{\partial}{\partial X} + \frac{\partial}{\partial T}\right) \left(\Psi^\prime e^{if}\right) = i\frac{\hbar}{2m} \frac{\partial^2}{\partial X^2}\left(\Psi^\prime e^{if}\right) + W^\prime\left(\Psi^\prime e^{if}\right)$.

If we were to take the derivatives in this equation, we would have to apply the product rule several times, because $f = f(X, T)$ We will see that this will in the end allow us to find an explicit expression for $f$. We will cut this differentiation into three pieces,

1: $\frac{\partial}{\partial X}\left(\Psi^\prime e^{if}\right)$ <br>

2: $\frac{\partial}{\partial T}\left(\Psi^\prime e^{if}\right)$ <br>

3: $\frac{\partial^2}{\partial X^2}\left(\Psi^\prime e^{if}\right)$


### Equation 1
This is simply done by applying the product rule,

$\frac{\partial}{\partial X}\left(\Psi^\prime e^{if}\right) = \frac{\partial \Psi^\prime}{\partial X}e^{if} + \Psi^\prime\frac{\partial e^{if}}{\partial X}$.

We cannot find a more explicit expression for $\frac{\partial \Psi^\prime}{\partial X}$ at this time, but we can still work out the second term a bit further, resulting in 

$\frac{\partial}{\partial X}\left(\Psi^\prime e^{if}\right) = \frac{\partial \Psi^\prime}{\partial X}e^{if} + i\frac{\partial f}{\partial X}\Psi^\prime e^{if}$.

We can take the exponential out of brackets to give

$\frac{\partial}{\partial X}\left(\Psi^\prime e^{if}\right) = \left(\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial f}{\partial X}\Psi^\prime\right) e^{if}$.


### Equation 2
This equation is similar to Equation 1. All we need to do is replace the Xs by Ts to get the result,

$\frac{\partial}{\partial T}\left(\Psi^\prime e^{if}\right) = \left(\frac{\partial \Psi^\prime}{\partial T} + i\frac{\partial f}{\partial T}\Psi^\prime\right) e^{if}$.


### Equation 3
This one is a bit more work, as we have to take a second derivative. We can start from Equation 1 and take an additional derivative to X,

$\frac{\partial^2}{\partial X^2}\left(\Psi^\prime e^{if}\right) = \frac{\partial}{\partial X} \left(\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial f}{\partial X}\Psi^\prime\right) e^{if}$.

This results in

$\frac{\partial^2}{\partial X^2}\left(\Psi^\prime e^{if}\right) = \left( \frac{\partial^2 \Psi^\prime}{\partial X^2}  + i\frac{\partial f}{\partial X}\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial^2 f}{\partial X^2} \Psi^\prime\right)e^{if} + i\frac{\partial f}{\partial X}\left(\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial f}{\partial X}\Psi^\prime\right)e^{if}$.

We can combine two terms to give us 

$\frac{\partial^2}{\partial X^2}\left(\Psi^\prime e^{if}\right) = \left( \frac{\partial^2 \Psi^\prime}{\partial X^2}  + 2i\frac{\partial f}{\partial X}\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial^2 f}{\partial X^2}\Psi^\prime  - \left(\frac{\partial f}{\partial X}\right)^2 \Psi^\prime \right)e^{if}$.

This looks very unwieldy, and it is, but things will soon simplify if we only push on a bit longer.

### Combining all Equations

We can plug Equation 1, 2 and 3 back in our original Schrödinger Equation. This gives us

$-v \left(\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial f}{\partial X}\Psi^\prime\right) e^{if} + \left(\frac{\partial \Psi^\prime}{\partial T} + i\frac{\partial f}{\partial T}\Psi^\prime\right) e^{if} = i\frac{\hbar}{2m} \left( \frac{\partial^2 \Psi^\prime}{\partial X^2}  + 2i\frac{\partial f}{\partial X}\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial^2 f}{\partial X^2}\Psi^\prime  - \left(\frac{\partial f}{\partial X}\right)^2 \Psi^\prime \right)e^{if} + W^\prime\Psi^\prime e^{if}$.

We can start our simplification by cancelling the exponentials, resulting in 

$-v \left(\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial f}{\partial X}\Psi^\prime\right) + \left(\frac{\partial \Psi^\prime}{\partial T} + i\frac{\partial f}{\partial T}\Psi^\prime\right) = i\frac{\hbar}{2m} \left( \frac{\partial^2 \Psi^\prime}{\partial X^2}  + 2i\frac{\partial f}{\partial X}\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial^2 f}{\partial X^2}\Psi^\prime  - \left(\frac{\partial f}{\partial X}\right)^2 \Psi^\prime \right) + W^\prime\Psi^\prime$.

We can now bring all terms to one side, giving us

$i\frac{\hbar}{2m} \left( \frac{\partial^2 \Psi^\prime}{\partial X^2}  + 2i\frac{\partial f}{\partial X}\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial^2 f}{\partial X^2}\Psi^\prime  - \left(\frac{\partial f}{\partial X}\right)^2 \Psi^\prime \right) + W^\prime\Psi^\prime + v \left(\frac{\partial \Psi^\prime}{\partial X} + i\frac{\partial f}{\partial X}\Psi^\prime\right) - \left(\frac{\partial \Psi^\prime}{\partial T} + i\frac{\partial f}{\partial T}\Psi^\prime\right) = 0$.

We will now group specific terms, giving

$\left(i\frac{\hbar}{2m} \frac{\partial^2 \Psi^\prime}{\partial X^2} + W^\prime\Psi^\prime - \frac{\partial \Psi^\prime}{\partial T}\right) + \frac{\partial \Psi^\prime}{\partial X}\left(-\frac{\hbar}{m}\frac{\partial f}{\partial X} + v \right) + \Psi^\prime \left(-\frac{\hbar}{2m} \frac{\partial^2 f}{\partial X^2} - i\frac{\hbar}{2m}\left(\frac{\partial f}{\partial X}\right)^2 +iv\frac{\partial f}{\partial X} - i\frac{\partial f}{\partial T}\right) = 0$.

It will soon become apparent why this grouping is effective. First of all, the first term is just the untransformed Schrödinger Equation in the moving frame's coordinates. Written this way, this term is zero, leaving us with

$\frac{\partial \Psi^\prime}{\partial X}\left(-\frac{\hbar}{m}\frac{\partial f}{\partial X} + v \right) + \Psi^\prime \left(-\frac{\hbar}{2m} \frac{\partial^2 f}{\partial X^2} - i\frac{\hbar}{2m}\left(\frac{\partial f}{\partial X}\right)^2 +iv\frac{\partial f}{\partial X} - i\frac{\partial f}{\partial T}\right) = 0$.

We now have two terms, grouped by the derivatives of $\Psi^\prime$. If this equation is to be zero for any value of $X$ and $T$, both terms between brackets must be zero. Thus, we have two equations,

4: $-\frac{\hbar}{m}\frac{\partial f}{\partial X} + v = 0$ <br>
5: $-\frac{\hbar}{2m} \frac{\partial^2 f}{\partial X^2} - i\frac{\hbar}{2m}\left(\frac{\partial f}{\partial X}\right)^2 + iv\frac{\partial f}{\partial X} - i\frac{\partial f}{\partial T} = 0$.

Both of these equations can be employed to find $f(X, T)$. Since the equations are formulated in terms of partial derivatives, we need both of them to solve the system entirely. We start with equation 4.

### Equation 4

We have 

$-\frac{\hbar}{m}\frac{\partial f}{\partial X} + v = 0$.

We can solve for $\frac{\partial f}{\partial X}$, giving us

$\frac{\partial f}{\partial X} = \frac{mv}{\hbar}$.

We can find $f$ by simply integrating both sides to $X$, giving

$f = \frac{mvX}{\hbar} + g(T)$.

The additional function $g(T)$ is present, because if we take the partial derivative to $X$, it drops out again and therefore needs to be included for now. We will employ Equation 5 to find $g(T)$.

### Equation 5

We have 

$-\frac{\hbar}{2m} \frac{\partial^2 f}{\partial X^2} - i\frac{\hbar}{2m}\left(\frac{\partial f}{\partial X}\right)^2 +iv\frac{\partial f}{\partial X} - i\frac{\partial f}{\partial T} = 0$.

We can group the imaginary terms, giving

$-\frac{\hbar}{2m} \frac{\partial^2 f}{\partial X^2} - i\left(\frac{\hbar}{2m}\left(\frac{\partial f}{\partial X}\right)^2 - v\frac{\partial f}{\partial X} + \frac{\partial f}{\partial T}\right) = 0$.

For this to be zero, both the real and imaginary parts need to be zero. We can start with the real part,

$\frac{\hbar}{2m} \frac{\partial^2 f}{\partial X^2} = 0$.

We can integrate this once to find

$\frac{\partial f}{\partial X} = h^\prime(T)$,

where $h^\prime(T)$ is a function that may depend on $T$. It can also be a constant, however. We can integrate to $X$ again to find

$f = h^\prime(T)X + h(T)$.

Again, $h(T)$ is another function of $T$, unrelated to $h^\prime(T)$, but I am running out of letters here. We can immediately compare this equation with our result from Equation 4. This tells us that 

$h^\prime(T) = \frac{mv}{\hbar}$,

which turned out to be independent of $T$ and thus

$h(T) = g(T)$.

Clearly, solving this part of the equation did not give us additional information. We can try solving the imaginary part,

$\frac{\hbar}{2m}\left(\frac{\partial f}{\partial X}\right)^2 - v\frac{\partial f}{\partial X} + \frac{\partial f}{\partial T} = 0$.

We already know from Equation 4 that 

$\frac{\partial f}{\partial X} = \frac{mv}{\hbar}$.

We can immediately plug this in to find

$\frac{\hbar}{2m}\left(\frac{mv}{\hbar}\right)^2 - v\frac{mv}{\hbar} + \frac{\partial f}{\partial T} = 0$.

Bringing $\frac{\partial f}{\partial T}$ to the LHS gives us

$\frac{\partial f}{\partial T} = \frac{mv^2}{2\hbar}$.

Integrating this equation gives us

$f = \frac{mv^2T}{2\hbar} + k(X)$,

where $K(X)$ is some function of $X$.

### Combining the Results

Combining the results from Equation 4 and 5, we can find our long-sought expression for $f$. It is readily seen that

$k(X) = \frac{mvX}{\hbar}$

and

$g(T) = \frac{mv^2T}{2\hbar}$.

This gives us the final result for $f$,

$f(X, T) = \frac{mvX}{\hbar} + \frac{mv^2T}{2\hbar}$.

This equation can be written more suggestively by noting that the kinetic energy and momentum are

$E = \frac{mv^2}{2}$ <br>
$p = mv$.

This allows us to write $f$ very succintly

$f(X, T) = \frac{pX + ET}{\hbar}$.

## Conclusion

Returning to earlier in this document, we showed that

$\Psi(x, t) = \Psi^\prime(X, T) e^{if(X, T)}$.

We can now plug in $f$ to find

$\Psi(x, t) = \Psi^\prime(X, T) e^{i\frac{pX + ET}{\hbar}}$.

We can now find our $\Psi(x, t)$ in our frame if we are given $Psi^\prime(X, T)$ in a moving frame. However, we would like to find $\Psi^\prime(X, T)$ from our original coordinates. We can simply take the exponential to the other side, giving us

$\Psi^\prime(X, T) = \Psi(x, t)e^{-i\frac{pX + ET}{\hbar}}$.

Finally, we would like to express the phase in terms of $x$ and $t$, rather than $X$ and $T$. We can return to our explicit expressions for the energy and momentum, giving us

$pX + ET = mvX + \frac{mv^2}{2}T$.

Plugging in the transformation equations for $X$ and $T$, we get

$mvX + \frac{mv^2}{2}T = mv\left(x - vt\right) + \frac{mv^2}{2}t = mvx - mv^2t + \frac{mv^2}{2}t = mvx - \frac{mv^2}{2}t = px - Et$.

Thus, the transformed wave function becomes 

$\Psi^\prime(X, T) = \Psi(x, t)e^{-i\frac{px - Et}{\hbar}}$.

It was a lot of work, but we arrived at a very general method for transforming wave equations in quantum mechanics. In the next document, we will apply this to the Gaussian wave packet.
